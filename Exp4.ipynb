# Importing Libraries

```python
import matplotlib.pyplot as plt
import numpy as np
from tensorflow.keras import models, datasets, layers, regularizers
from tensorflow.keras.initializers import GlorotNormal, HeNormal
```

# Dataset loading and Preparation

```python
(x_train, y_train), (x_test, y_test)= datasets.cifar10.load_data()
```

```python
x_train=x_train/255.0
x_test=x_test/255.0
```

# Model Building

### 1. Base Model

```python
b_model= models.Sequential([
    layers.Flatten(input_shape=(32,32,3)),
    layers.Dense(128, activation='relu'),
    layers.Dense(64, activation='relu'),
    layers.Dense(32, activation='relu'),
    layers.Dense(10, activation='softmax')
])
```

### 2. Xavier model

```python
x_model=models.Sequential([
    layers.Flatten(input_shape=(32,32,3)),
    layers.Dense(128, activation='relu', kernel_initializer=GlorotNormal()),
    layers.Dense(64, activation='relu', kernel_initializer=GlorotNormal()),
    layers.Dense(32, activation='relu', kernel_initializer=GlorotNormal()),
    layers.Dense(10, activation='softmax')
])
```

### 3. Kaming Model

```python
k_model=models.Sequential([
    layers.Flatten(input_shape=(32,32,3)),
    layers.Dense(128, activation='relu', kernel_initializer=HeNormal()),
    layers.Dense(64, activation='relu', kernel_initializer=HeNormal()),
    layers.Dense(32, activation='relu', kernel_initializer=HeNormal()),
    layers.Dense(10, activation='softmax'),

])
```

### 4. Dropout Model

```python
model=models.Sequential([
    layers.Flatten(input_shape=(32,32,3)),
    layers.Dense(128, activation='relu'),
    layers.Dropout(0.1),
    layers.Dense(64, activation='relu'),
    layers.Dropout(0.1),
    layers.Dense(32, activation='relu'),
    layers.Dropout(0.1),
    layers.Dense(10, activation='softmax')
])
```

### L2 regularized Model

```python
l2_model=models.Sequential([
    layers.Flatten(input_shape=(32,32,3)),
    layers.Dense(128, activation='relu', kernel_regularizer=regularizers.l2(0.01)),
    layers.Dense(64, activation='relu', kernel_regularizer=regularizers.l2(0.01)),
    layers.Dense(32, activation='relu', kernel_regularizer=regularizers.l2(0.01)),
    layers.Dense(10, activation='softmax')
])
```

### Compiling and Fitting

```python
b_model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])
x_model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])
k_model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])
model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])
l2_model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])
```

#### 1. Base Model

```python
r_b=b_model.fit(x_train, y_train, epochs=50, validation_data=(x_test, y_test))
```

#### 2. Xavier Model

```python
r_x=x_model.fit(x_train, y_train, epochs=50, validation_data=(x_test, y_test))
```

#### 3. Kaming Model

```python

r_k=k_model.fit(x_train, y_train, epochs=50, validation_data=(x_test, y_test))
```

#### 4. Dropout Model

```python

r=model.fit(x_train, y_train, epochs=50, validation_data=(x_test, y_test))
```

#### 5. L2 Model

```python

r_l2=l2_model.fit(x_train, y_train, epochs=50, validation_data=(x_test, y_test))
```
